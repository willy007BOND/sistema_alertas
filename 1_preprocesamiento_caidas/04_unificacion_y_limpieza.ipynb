# 3. Unificación de Carpetas YOLO (Cleanup)

Este notebook fusiona la carpeta antigua de `TFM/YOLO_DATASET_UNIFIED` con la nueva en la raíz del SSD, asegurando que no haya duplicidad.

**Entorno Recomendado:** `tfmvc`

```python
import shutil
import os
from pathlib import Path
from tqdm import tqdm

# Rutas
SOURCE = Path("/Volumes/Echo 13 SSD/TFM/YOLO_DATASET_UNIFIED")
TARGET = Path("/Volumes/Echo 13 SSD/YOLO_DATASET_UNIFIED")

def merge_datasets():
    if not SOURCE.exists():
        print("La carpeta de origen ya no existe. Unificación previamente completada.")
        return
        
    subdirs = ["images/train", "images/val", "labels/train", "labels/val"]

    for sd in subdirs:
        src_path = SOURCE / sd
        dst_path = TARGET / sd
        if not src_path.exists(): continue
        dst_path.mkdir(parents=True, exist_ok=True)
        
        files = os.listdir(src_path)
        for f in tqdm(files, desc=f"Merging {sd}"):
            if not (dst_path / f).exists():
                shutil.copy2(src_path / f, dst_path / f)

if __name__ == "__main__":
    merge_datasets()
    print("Unificación completada.")
```
