# 2. Centralización de Dataset KUMAR a Formato YOLO

Este notebook centraliza el dataset Kumar (que ya está en formato YOLO) en la carpeta unificada, añadiendo el prefijo `KUMAR_` para evitar colisiones.

**Entorno Recomendado:** `tfmvc`

```python
import shutil
from pathlib import Path
from tqdm import tqdm

# Rutas
SOURCE = Path("/Volumes/Echo 13 SSD/TFM/images/CAIDAS/KUMAR_FALL_DETECTION/dataset")
TARGET = Path("/Volumes/Echo 13 SSD/YOLO_DATASET_UNIFIED")

def centralize():
    for split in ["train", "val"]:
        img_dst = TARGET / "images" / split
        lbl_dst = TARGET / "labels" / split
        img_dst.mkdir(parents=True, exist_ok=True)
        lbl_dst.mkdir(parents=True, exist_ok=True)
        
        img_src = SOURCE / split
        lbl_src = SOURCE / "labels" / split
        
        if not img_src.exists(): continue

        images = list(img_src.glob("*.jpg"))
        for img_path in tqdm(images, desc=f"Copiando Kumar {split}"):
            new_name = f"KUMAR_{img_path.name}"
            shutil.copy2(img_path, img_dst / new_name)
            
            lbl_path = lbl_src / f"{img_path.stem}.txt"
            if lbl_path.exists():
                shutil.copy2(lbl_path, lbl_dst / f"KUMAR_{lbl_path.name}")

if __name__ == "__main__":
    centralize()
    print("Centralización de Kumar finalizada.")
```
