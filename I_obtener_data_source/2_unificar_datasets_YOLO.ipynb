# Unificación de Datasets YOLO

Este notebook organiza y unifica datasets en formato YOLO (imágenes y etiquetas divididas en `train` y `val`) desde múltiples fuentes en una única carpeta consolidada.

### Características:
- **Prefijos de Seguridad:** Añade el nombre del dataset al nombre del archivo para evitar colisiones.
- **Estructura YOLO:** Mantiene la jerarquía `images/train`, `images/val`, `labels/train`, `labels/val`.
- **Preservación:** Realiza copias (`shutil.copy2`) para no modificar los archivos originales.

```python
import os
import shutil
from pathlib import Path

# Configuración de rutas
SOURCE_BASE = "/Volumes/Echo 13 SSD/TFM/images/CAIDAS"
TARGET_BASE = "/Volumes/Echo 13 SSD/YOLO_DATASET_UNIFIED"

# Lista de datasets detectados con estructura YOLO
DATASETS = [
    {
        "name": "FALL_DETECTION",
        "images": "FALL_DETECTION/fall_dataset/fall_dataset/images",
        "labels": "FALL_DETECTION/fall_dataset/fall_dataset/labels"
    },
    {
        "name": "KUMAR_FALL_DETECTION",
        "images": "KUMAR_FALL_DETECTION/dataset",
        "labels": "KUMAR_FALL_DETECTION/dataset/labels"
    }
]

def setup_structure(base_path):
    """Crea la estructura de carpetas YOLO si no existe."""
    paths = [
        "images/train", "images/val",
        "labels/train", "labels/val"
    ]
    for p in paths:
        os.makedirs(os.path.join(base_path, p), exist_ok=True)
    print(f"Estructura preparada en: {base_path}")

def copy_files(ds_name, src_dir, target_subdir, prefix_type):
    """Copia archivos añadiendo un prefijo para evitar duplicados."""
    if not os.path.exists(src_dir):
        print(f"  [!] Aviso: {src_dir} no existe. Saltando.")
        return 0
    
    count = 0
    for filename in os.listdir(src_dir):
        # Ignorar archivos ocultos del sistema
        if filename.startswith('.'): continue
        
        src_path = os.path.join(src_dir, filename)
        if os.path.isdir(src_path): continue
        
        # Nuevo nombre con prefijo del dataset
        new_filename = f"{ds_name}_{filename}"
        dest_path = os.path.join(TARGET_BASE, prefix_type, target_subdir, new_filename)
        
        shutil.copy2(src_path, dest_path)
        count += 1
    return count
```

```python
# Ejecución del proceso
setup_structure(TARGET_BASE)

for ds in DATASETS:
    name = ds["name"]
    print(f"\nProcesando dataset: {name}")
    
    for split in ["train", "val"]:
        # Procesar Imágenes
        src_img_dir = os.path.join(SOURCE_BASE, ds["images"], split)
        img_count = copy_files(name, src_img_dir, split, "images")
        
        # Procesar Etiquetas
        src_lbl_dir = os.path.join(SOURCE_BASE, ds["labels"], split)
        lbl_count = copy_files(name, src_lbl_dir, split, "labels")
        
        print(f"  - {split}: {img_count} imágenes, {lbl_count} etiquetas copiadas.")

print("\n¡Proceso finalizado con éxito!")
```
